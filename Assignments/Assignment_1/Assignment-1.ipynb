## Q7. Temperature Converter

Convert between Celsius and Fahrenheit using user input and type conversion.

```python
ask = input("Type = ['c' for Celius and 'f' for Fahrenheit]")
num = float(input("Enter your number"))

if ask == "c":
    print("Temp in Fehrenheit is:", (num * (9/5)) + 32,"F")
elif ask == "f":
    print("Temp in Celsius is:", (num - 32) * (5/9), "C")
else:
    print("Invailed Request")  
```

## Q8. Variables & Mutability

Explain mutable vs immutable with code showing list (mutable) and tuple (immutable) behavior.

```python
## list is mutable so we can add remove or replace values

lis = [1,2,3,4,"a","b","c","d","!","@"]

lis += [8]                                 ## when you want to add Value in List
lis [2] = 5                                ## when you want to replace the value in list
lis.remove(1)                              ## when you want to remove the value in list
print(lis)

## Tuple is immutable so we can not add remove or replace any value it create new tuple

tupl = (1,2,3,4,"a","b","c","d","!","@")
tupl (2) = 5
print(tupl)
```

## Q10. Keywords & Identifiers

Explain valid Python identifiers and write code to check invalid ones or keywords.
#### Answer: 

A valid Python identifier is a name for a variable, function, class, or module they all follows default rules for example:
1. It must start with a letter (a-z, A-Z) or an underscore (_), you can use numbers (0-9) in between or last.
2. They can not start with any special characters (like @,#).
3. They can not contain space in between the words.
4. They can not use Reserved Keywords as a variable name.
##### Example:
1. CAMEL - kidsAge,
2. PASCAL - KidsAge,
3. SNAKE - kids_age

```python
## Vaild 

kidsAge = "16"     # CAMEL
KidsAge = "17"     # PASCAL
kids_age = "18"    # SNAKE
_kidsage = "19"
_kids_age = "20"
kidsage09 = "21"
kids09age = "22"

print(kidsAge)
print(KidsAge)
print(kids_age)
print(_kidsage)
print(_kids_age)
print(kidsage09)
print(kids09age)

```

```python
## Invaild

@kidsAge = "16"       ## starts with special character
kids age = "17"       ## space in between words
while = "18"          ## Reserved Keyword
as = "19"             ## Reserved Keyword
9kidsage = "20"       ## start with number

print(@kidsAge)
print(kids age)
print(while)
print(as)
print(9kidsage)
```
