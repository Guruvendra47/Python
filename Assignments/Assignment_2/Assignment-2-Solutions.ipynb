## Q5. Leap Year Checker (Nested if)

Write a program to determine whether a given year is a leap year using nested conditions.

```python
year= int(input("Enter the year"))

## It is a nested condition. If the year is divisible by 4 and the remainder is 0,
## then the program goes inside this block.
## If the year is not divisible by 4 (remainder is not 0), then it goes to the else block
## and it is not a leap year.

if year % 4 == 0:
    if year % 100 != 0:
        if year % 400 != 0:
            print("Leap year but not century Year")
    elif year % 100 == 0:              ## Since we have already checked year % 4 == 0 in the outer if,                        
        if year % 400 != 0:            ## we only need nested if/else inside it — no separate elif at the same level.
            print("Not a Leap year it is a century year but not divisible by 400")
        elif year % 400 == 0:
            print("Leap year and also rare century Year that is divisible by 400")

else:
    print("Not Leap Year")
    
```
