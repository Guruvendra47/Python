## 1. To print Given Number in Column

```python
n = int(input("Enter your Number"))

for i in range(n):
        print("* ")
```

### 1.1. To print Given Number in Row
By default, the print() function automatically adds a newline character '(\n)' to the end of the line, making subsequent print statements appear on new lines.

The 'end =' parameter allows you to override this default behavior and you can specify an alternative ending, such as a space, a comma, an exclamation mark, or an empty string, to control the output format.

```python
n = int(input("Enter your Number"))

for i in range(n):
        print("* ", end = " ")  ## end = allows to override the default behavior print characters in single row.
```

## 10. Top Half Hallow Diamond Pattern with *

```python
n = int(input("Enter your Number"))

for i in range(n):
    print("  "*(n-i-1)+"*", end=" ")
    if i >=1:
        print("  "*(2*i-1)+"*", end=" ")
    print()
```

## 11. Bottom Half Hallow Diamond Pattern with *¶

```python
n=int(input("Enter your Number"))

for i in range(n):
    print("  "*(i)+"*", end=" ")
    if i != (n-1):
        print("  "*(2*n-2*i-3)+"*", end=" ")
    print()
```
